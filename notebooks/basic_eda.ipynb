```python
def get_hemorrhage_mask(filename):
    filename = Path(filename)
    mask_filename = filename.parent / (filename.stem + '_HGE_Seg.jpg')
    return mask_filename if mask_filename.exists() else None
```

```python
import matplotlib.pyplot as plt

def display_images(row, show_header_info=False):
    'requires a `row` from the `hemorrhages_df` dataframe and will print patient and lesion information while showing brain, bone, and lesion images side-by-side'
    if show_header_info:
        print(row)
        print(patients_df[patients_df.index == row.PatientNumber].iloc[0])
    brain_display_filename = row['brain filename']
    bone_display_filename = row['bone filename']
    hemorrhage_mask_filename = get_hemorrhage_mask(brain_display_filename)

    brain_img = plt.imread(brain_display_filename)
    bone_img = plt.imread(bone_display_filename)

    f, axs = plt.subplots(1,3, dpi=300)
    axs[0].axis('off')
    axs[0].imshow(bone_img, cmap='gray')
    axs[0].set_title('bone ww/wl')
    axs[1].axis('off')
    axs[1].imshow(brain_img, cmap='gray')
    axs[1].set_title('brain ww/wl')

    if hemorrhage_mask_filename:
        mask_img = plt.imread(hemorrhage_mask_filename)
        axs[2].axis('off')
        axs[2].imshow(brain_img, cmap='gray')
        axs[2].imshow(mask_img, cmap='Reds', alpha=0.5)
        axs[2].set_title('hemorrhage overlay')

row = hemorrhages_df[hemorrhages_df['No_Hemorrhage'] == False].iloc[0]
display_images(row, show_header_info=True)
```

Now write a quick function to measure the hemorrhage pixels for each row in the `hemorrhage dataframe`

```python
def measure_slicewise_hemorrhage_volume(fname):
    threshold = 100
    lesion_mask_fname = get_hemorrhage_mask(fname)
    if lesion_mask_fname:
        lesion_mask = plt.imread(lesion_mask_fname) > threshold
        slice_vol = lesion_mask.sum()
    else:
        slice_vol = 0
    return slice_vol
```

```python
hemorrhages_df['hemorrhage volume [pixels]'] = hemorrhages_df['brain filename'].apply(measure_slicewise_hemorrhage_volume)
hemorrhages_df.head()
```

According to the dataset paper [Hssayeni et al 2020](https://www.mdpi.com/2306-5729/5/1/14) scans had an inplace resolution of 0.33 mm and 5 mm slice thickness, we can use this to convert from pixel volume to mm^3 and mL volume

```python
slice_thickness_mm = 5
pixel_size_mm = 0.33
mm3_per_ml = 1000
hemorrhages_df['hemorrhage volume [mL]'] = hemorrhages_df['hemorrhage volume [pixels]'].apply(lambda o: o*slice_thickness_mm*pixel_size_mm**2/mm3_per_ml)
```

## Hemorrhage Density

Here's we'll be using both the mask and the images to look at the voxel intensity histograms inside the segmented hemorrhages...

Try to look into quantitative descriptors of heterogeneity to capture the irregular density such as [entropy](https://scikit-image.org/docs/stable/auto_examples/filters/plot_entropy.html)

```python
def get_hemorrhage_intensity_histogram(img_fname):
    mask_fname = get_hemorrhage_mask(img_fname)
    img, mask = plt.imread(img_fname), plt.imread(mask_fname)
    histogram, bin_edges = np.histogram(img[mask > threshold], bins=256, range=(0, 256), density=True)
    return histogram, bin_edges[:-1]
```

```python
from scipy.stats import skew, kurtosis
def get_hemorrhage_intensity_mean(img_fname):
    mask_fname = get_hemorrhage_mask(img_fname)
    if mask_fname is None:
        return 0
    img, mask = plt.imread(img_fname), plt.imread(mask_fname)
    return img[mask > threshold].mean()

def get_hemorrhage_intensity_std(img_fname):
    mask_fname = get_hemorrhage_mask(img_fname)
    if mask_fname is None:
        return 0
    img, mask = plt.imread(img_fname), plt.imread(mask_fname)
    return img[mask > threshold].std()

def get_hemorrhage_intensity_skew(img_fname):
    mask_fname = get_hemorrhage_mask(img_fname)
    if mask_fname is None:
        return 0
    img, mask = plt.imread(img_fname), plt.imread(mask_fname)
    return skew(img[mask > threshold])

def get_hemorrhage_intensity_kurtosis(img_fname):
    mask_fname = get_hemorrhage_mask(img_fname)
    if mask_fname is None:
        return 0
    img, mask = plt.imread(img_fname), plt.imread(mask_fname)
    return kurtosis(img[mask > threshold])
```

```python
hemorrhages_df['hemorrhage intensity mean'] = hemorrhages_df['brain filename'].apply(get_hemorrhage_intensity_mean)
hemorrhages_df['hemorrhage intensity std'] = hemorrhages_df['brain filename'].apply(get_hemorrhage_intensity_std)
hemorrhages_df['hemorrhage intensity skew'] = hemorrhages_df['brain filename'].apply(get_hemorrhage_intensity_skew)
hemorrhages_df['hemorrhage intensity kurtosis'] = hemorrhages_df['brain filename'].apply(get_hemorrhage_intensity_kurtosis)
hemorrhages_df
```

```python
def pediatric_subgroup(age):
    if age < 1:
        return 'newborn'
    elif (age >= 1) & (age < 5):
        return 'infant'
    elif (age >= 5) & (age < 12):
        return 'child'
    elif (age >= 12) & (age < 22):
        return 'adolescent'
    else:
        return 'adult'

patients_df['pediatric'] = patients_df['Age\n(years)'].apply(lambda o: o<22).astype('category')
patients_df['pediatric subgroup'] = patients_df['Age\n(years)'].apply(pediatric_subgroup).astype('category')
```

```python
def plot_intensity_histogram(img_fname):
    threshold = 100
    mask_fname = get_hemorrhage_mask(img_fname)
    if mask_fname:
            histogram, bins = get_hemorrhage_intensity_histogram(img_fname)
    img = plt.imread(img_fname)
    mask = plt.imread(mask_fname) if mask_fname else None
    
    f, axs = plt.subplots(2,3, tight_layout=True, figsize=(12,8))
    axs = axs.flatten()
    axs[0].imshow(img, cmap='gray')
    axs[0].axis('off')
    axs[0].set_title('hemorrhage overlay')
    
    axs[3].imshow(img, cmap='gray')
    if mask_fname:
        axs[3].imshow(mask, alpha=0.5, cmap='Reds')
        axs[3].axis('off')
    
    axs[3].set_title('hemorrhage overlay')
    if mask_fname:
        axins = axs[3].inset_axes([0.6, 0.6, 0.35, 0.35])
        axins.plot(bins, histogram, alpha=0.5)
        
        mean = img[mask > threshold].mean()
        std = img[mask > threshold].std()
        
        axins.set_title(f'mean: {mean:3.0f} | std: {std:3.0f}', fontsize=8)
        axins.set_ylabel('frequency', fontsize=8)
        axins.set_xlabel('gray value', fontsize=8)
    
    pd.DataFrame({'measure': hemorrhages_df[hemorrhages_df['brain filename'] == img_fname]['hemorrhage intensity mean'], 'dataset mean': hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity mean'].mean()}).plot(ax=axs[2], kind='bar')
    axs[2].set_title('pixel intensity mean')
    axs[2].set_ylim([hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity mean'].min(),  hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity mean'].max()])
    
    pd.DataFrame({'measure': hemorrhages_df[hemorrhages_df['brain filename'] == img_fname]['hemorrhage intensity std'], 'dataset mean': hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity std'].mean()}).plot(ax=axs[1], kind='bar')
    axs[1].set_title('pixel intensity std')
    axs[1].set_ylim([ hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity std'].min(),  hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity std'].max()])

    
    pd.DataFrame({'measure': hemorrhages_df[hemorrhages_df['brain filename'] == img_fname]['hemorrhage intensity skew'], 'dataset mean': hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity skew'].mean()}).plot(ax=axs[4], kind='bar')
    axs[4].set_title('pixel intensity skew')
    axs[4].set_ylim([ hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity skew'].min(),  hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity skew'].max()])
    
    pd.DataFrame({'measure': hemorrhages_df[hemorrhages_df['brain filename'] == img_fname]['hemorrhage intensity kurtosis'], 'dataset mean': hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity kurtosis'].mean()}).plot(ax=axs[5], kind='bar')
    axs[5].set_title('pixel intensity kurtosis')
    axs[5].set_ylim([ hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity kurtosis'].min(),  hemorrhages_df[hemorrhages_df['hemorrhage volume [mL]'] > 0]['hemorrhage intensity kurtosis'].max()])
```

```python
def display_intensity_histogram_widget(hemorrhage=True, patientnumber=49, image_index=0, show_header_info=False):
    'note that `image_index` does not necesarrily agree with slice number since `image_index` starts at the bottom of the scan if there is not hemorrhage or at the first segmentation slice if a segmentation is present'
    filtered_df = hemorrhages_df[(hemorrhages_df['No_Hemorrhage'] != hemorrhage) & (hemorrhages_df['PatientNumber']==patientnumber)]
    if image_index > len(filtered_df):
        print(f'image_index: {image_index} exceeds the available slices: {len(filtered_df)}, please select a smaller index to view')
    row = filtered_df.sort_values(by='SliceNumber').iloc[image_index]
    if show_header_info:
        print(row)
    plot_intensity_histogram(row['brain filename'])

max_slice_idx = max([len(hemorrhages_df[(hemorrhages_df.PatientNumber == n)]) for n in hemorrhages_df.PatientNumber.unique()])
    
interact(display_intensity_histogram_widget,
         hemorrhage=[True, False],
         patientnumber=patients_df.index.unique(),
         image_index=IntSlider(value=0, min=0, max=max_slice_idx),
         show_header_info=Checkbox(value=False, description='print header'))
```
